```python
import pandas as pd

df = spark.table("ml_perovskites.db_ml").toPandas()

# normalize column names
df.columns = df.columns.str.lower().str.strip()

# normalize reaction_type
df["reaction_type"] = df["reaction_type"].str.strip().str.lower()

# keep only photoreforming
df = df[df["reaction_type"] == "photoreform"]

# clean compound formatting
df["compound"] = df["compound"].str.replace(" ", "")

# numeric conversion
numeric_cols = [
    "hydrogen_production_rate",
    "bandgap_energy",
    "particle_size",
    "radiation_source_intensity",
    "specific_surface_area",
    "catalyst_amount",
    "calcination_temperature",
    "calcination_time"
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

# remove missing key values
df = df.dropna(subset=["compound", "hydrogen_production_rate"])

# remove impossible values
df = df[df["hydrogen_production_rate"] >= 0]
df = df[df["bandgap_energy"] > 0]

# export
df.to_csv("clean_dataset_db_ml.csv", index=False)
```
