```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("composition_dataset_db_ml.csv")

# detect element columns
element_cols = [c for c in df.columns if c[0].isupper()]

# count compounds containing each element
element_counts = (df[element_cols] > 0).sum()

# sort
element_counts = element_counts.sort_values(ascending=False)

# normalize values for colormap
norm = plt.Normalize(element_counts.min(), element_counts.max())
colors = plt.cm.viridis(norm(element_counts.values))

plt.figure(figsize=(16,6))

bars = plt.bar(element_counts.index, element_counts.values, color=colors)

plt.ylabel("Element Count")
plt.xlabel("Element")
plt.title("Element prevalence in dataset")

# add numbers above bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1,
        int(height),
        ha='center',
        va='bottom',
        fontsize=8
    )

plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import pymatviz as pmv

df = pd.read_csv("composition_dataset_db_ml.csv")

# detect element columns
element_cols = [c for c in df.columns if c[0].isupper()]

# count compounds containing each element
element_counts = (df[element_cols] > 0).sum()

# create figure
fig = pmv.ptable_heatmap_plotly(element_counts.to_dict())

# change background
fig.update_layout(
    paper_bgcolor="white",
    plot_bgcolor="white"
)

fig.show()
```
