```python
import pandas as pd
from pymatgen.core import Composition

INPUT_FILE = "clean_dataset_db_ml.csv"
OUTPUT_FILE = "composition_dataset_db_ml.csv"
FORMULA_COLUMN = "compound"


def parse_composition(formula):
    try:
        comp = Composition(formula)
        return comp.get_el_amt_dict()
    except Exception:
        return {}


def expand_compositions(df, formula_col):

    parsed = df[formula_col].apply(parse_composition)

    elements_df = pd.json_normalize(parsed)

    elements_df = elements_df.fillna(0)

    df_out = pd.concat([df, elements_df], axis=1)

    return df_out


def main():

    df = pd.read_csv(INPUT_FILE)

    df = expand_compositions(df, FORMULA_COLUMN)

    df.to_csv(OUTPUT_FILE, index=False)

    print("Composition parsing finished.")
    print(f"Output file: {OUTPUT_FILE}")


if __name__ == "__main__":
    main()
```
